```python

import linecache

fileA = open('mat1.txt', 'r' )
print(fileA.read())
print(" ")
line1 = linecache.getline('mat1.txt', 1)
line2 = linecache.getline('mat1.txt', 2)
line3 = linecache.getline('mat1.txt', 3)

one = line1.split()
list1 = list(map(int, one))

two = line2.split()
list2 = list(map(int,two))

three = line3.split()
list3 = list(map(int, three))

listA = [list1]+[list2]+[list3]
print (listA)
print(" ")


fileB = open('mat2.txt', 'r')
print(fileB.read())
print(" ")
Bline1 = linecache.getline('mat2.txt', 1)
Bline2 = linecache.getline('mat2.txt', 2)
Bline3 = linecache.getline('mat2.txt', 3)

bone = Bline1.split()
blist1 = list(map(int, bone))

btwo = Bline2.split()
blist2 = list(map(int,btwo))

bthree = Bline3.split()
blist3 = list(map(int, bthree))

listB = [blist1] + [blist2] + [blist3]
print (listB)
print(" ")

res1 = [[0] * len(listB[0]) for _ in range(len(listA))]
for i in range(len(listA)):
    for j in range(len(listB[0])):
        for k in range(len(listB)):
            res1[i][j] += listA[i][k]*listB[k][j]
print(res1)
print(" ")

def tvec(X):
    return[[i] for i in X]
X=[3,5.5,-9]
tv=tvec(X)
print(tv)
print(" ")
res2=[[0] for _ in range(len(listA))]
for i in range(len(listA)):
    for j in range(len(tv[0])):
        for k in range (len(tv)):
            res2[i][j]+=listA[i][k]*tv[k][j]
print(res2)
```
